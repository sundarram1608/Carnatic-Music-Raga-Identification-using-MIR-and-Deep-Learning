### 2. Features Transformation for CNN & LSTM

#### Build raw lists for CNN + LSTM & Collect CNN features like raw spectral matrices per clip & RNN features like pitch+gamaka sequences only for valid_indices

This is possible by collecting the both together

***CNN***
> Constructing a 2D “image” per clip by stacking vertically. (Here, image is a 2D matrix. CNN operates on images represented as 2D matrices with pixel information and hence this metaphor).

> Time dimension T is the number of frames. Since clips are all ~30 seconds and hop_length is consistent, T should be similar but not always identical → we will pad to max T.

***LSTM***
> We’ll create a time × feature matrix per clip, then pad.

> For each clip, we’ll use: f0_cents,interval_sequence, f0_derivative,f0_second_derivative,gamaka_extent,gamaka_rate,modulation_index.

> So each time step will have a 7D feature vector

```python
df = raga_features_merged_df.copy()

cnn_raw_list = []
lstm_raw_list = []
label_raw_list = []

for idx, row in df.iterrows():
    try:
        # ===== CNN FEATURES =====
        mel = row["mel_spectrogram_128"]
        mfcc = row["mfcc"]
        chroma = row["chroma_12"]

        T_mel = mel.shape[1]
        T_mfcc = mfcc.shape[1]
        T_chroma = chroma.shape[1]

        if not (T_mel == T_mfcc == T_chroma):
            print(f"Skipping idx {idx}: time mismatch")
            continue

        cnn_tensor = np.vstack([mel, mfcc, chroma])   # shape (153, T)

        # ===== LSTM FEATURES =====
        f0_cents = np.array(row["f0_cents"], dtype=float)
        interval_seq = np.array(row["interval_sequence"], dtype=float)
        f0_der = np.array(row["f0_derivative"], dtype=float)
        f0_second_der = np.array(row["f0_second_derivative"], dtype=float)
        g_extent = np.array(row["gamaka_extent"], dtype=float)
        g_rate = np.array(row["gamaka_rate"], dtype=float)
        g_mod = np.array(row["modulation_index"], dtype=float)

        N = len(f0_cents)
        if N < 5:
            print(f"Skipping idx {idx}: too short ({N})")
            continue

        T_pitch = N - 3
        seq = np.zeros((T_pitch, 7), dtype=np.float32)

        for t in range(2, N-1):
            k = t - 2
            seq[k, 0] = f0_cents[t]
            seq[k, 1] = interval_seq[t-1] if len(interval_seq) >= t else 0
            seq[k, 2] = f0_der[t-1] if len(f0_der) >= t else 0
            seq[k, 3] = f0_second_der[t-2] if len(f0_second_der) >= (t-1) else 0
            seq[k, 4] = g_extent[t] if len(g_extent) > t else 0
            seq[k, 5] = g_rate[t] if len(g_rate) > t else 0
            seq[k, 6] = g_mod[t] if len(g_mod) > t else 0

        seq = np.nan_to_num(seq, nan=0.0)

        # If both CNN + LSTM succeeded
        cnn_raw_list.append(cnn_tensor)
        lstm_raw_list.append(seq)
        label_raw_list.append(row["raga"])

    except Exception as e:
        print("Skipping idx", idx, "due to error:", e)
```

### 4. Padding the features & Encoding the target variables (One hot encoding)

#### 4.1 Padding functions for CNN & LSTM

```python
def pad_cnn_list(raw_list, freq_dim=153):
    T_max = max(arr.shape[1] for arr in raw_list)
    N = len(raw_list)
    out = np.zeros((N, freq_dim, T_max, 1), dtype=np.float32)

    for i, arr in enumerate(raw_list):
        T_i = arr.shape[1]
        out[i, :, :T_i, 0] = arr

    return out


def pad_lstm_list(raw_list, num_features=7):
    T_max = max(arr.shape[0] for arr in raw_list)
    N = len(raw_list)
    out = np.zeros((N, T_max, num_features), dtype=np.float32)

    for i, arr in enumerate(raw_list):
        T_i = arr.shape[0]
        out[i, :T_i, :] = arr

    return out
```

## B. CNN Branch (Spectral side)

### 1. About CNN Branch

This branch consumes the padded (153 × T_spec_max × 1) tensors below, stacked vertically into 153 frequency bands.
> 128 Mel bins

> 13 MFCCs

> 12 Chromas

***What this CNN Branch Learns***<br><br>
✔ Mel-spectrogram (128 channels)<br>
Timbre<br>
Texture<br>
Harmonic envelope<br>
Formants<br>
Brightness / nasality<br><br>
✔ MFCC (13 channels)<br>
Vocal tract shape<br>
Tone quality (breathy vs tense)<br>
Broad raga identity cues<br><br>
✔ Chroma (12 channels)<br>
Swaras used<br>
Notes prominence<br>
Raga scale structure<br><br>
✔ CNN learns:<br>
local spectral patterns<br>
gamaka textures (like glide fingerprints)<br>
energy distributions<br>
transition regions<br><br>
***Final CNN embedding is a 256-dimensional vector.***


### 2. UDF for CNN Archtecture building

```python
def build_cnn_branch(input_shape,cnn_filters=[32, 64, 128],kernel_sizes=[(3,3), (3,3), (3,3)],pool_sizes=[(2,2), (2,2), (2,2)],dropout_rate=0.3):

    """
    CNN Branch to process spectral features:
      - Mel-Spectrogram (128 bins)
      - MFCC (13 bins)
      - Chroma (12 bins)
      Combined → (153 × T × 1)
    """

    cnn_input = layers.Input(shape=input_shape, name="cnn_input")

    x = cnn_input

    # Block 1
    x = layers.Conv2D(cnn_filters[0], kernel_sizes[0], padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(pool_sizes[0])(x)

    # Block 2
    x = layers.Conv2D(cnn_filters[1], kernel_sizes[1], padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(pool_sizes[1])(x)

    # Block 3
    x = layers.Conv2D(cnn_filters[2], kernel_sizes[2], padding="same", activation="relu")(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling2D(pool_sizes[2])(x)

    # Flatten and Dense compression (CNN → embedding)
    x = layers.Flatten()(x)
    x = layers.Dropout(dropout_rate)(x)
    cnn_embedding = layers.Dense(256, activation="relu", name="cnn_embedding")(x)

    return cnn_input, cnn_embedding
```

## C. LSTM Branch (Pitch side)

### 1. About LSTM Branch


This branch consumes padded LSTM input, where each time step contains the 7 raga-critical features:

| Feature              | Meaning                         |
| -------------------- | ------------------------------- |
| f0_cents             | melodic contour                 |
| interval_sequence    | swara steps                     |
| f0_derivative        | slope → gamaka speed            |
| f0_second_derivative | curvature → gamaka acceleration |
| gamaka_extent        | oscillation depth               |
| gamaka_rate          | oscillation frequency           |
| modulation_index     | strength of the gamaka          |


What this LSTM Branch Learns<br><br>
✔ Bi-directionality<br>
The model sees both forward and backward temporal dependencies — extremely important for raga phrases (prayogas).<br><br>
✔ Learns pitch movements such as:<br>
Sa→Ri→Ga → smooth ascent<br>
Ni↓Dha↓Pa → descent<br>
Kampita (oscillation)<br>
Jaaru (glides)<br>
Stress patterns for nyasa swaras<br>
Characteristic bends (e.g., Kalyani’s N→S)<br><br>
✔ Learns gamaka properties:<br>
how deep the oscillations are<br>
how fast they vibrate<br>
how quickly pitch accelerates and de-accelerates<br>
the modulation strength<br><br>
✔ The branch outputs a 128-dimensional LSTM embedding.

### 2. UDF for LSTM Model building

```python
def build_lstm_branch(input_shape,lstm_units=128,dropout_rate=0.3):#,use_masking=True):

    """
    LSTM Branch for pitch + gamaka sequences.

    Input is (T, 7) where each timestep contains:
        [f0_cents, interval, df0, d2f0, g_extent, g_rate, g_mod]

    BiLSTM learns:
        - melodic contour patterns
        - characteristic prayogas
        - gamaka movement shape (kampita, jaaru)
        - swara transitions and raga grammar
    """

    lstm_input = layers.Input(shape=input_shape, name="lstm_input")
    x = lstm_input


    # masking (ignores padded zeros)
    # ❗ No Masking - we disable it due to internal zeros in features
    ''' “Ignore all padded timesteps (zeros).
        They are NOT real musical information.”
        Masking ensures:
          these zeros aren't treated as “silent frames”
          they don't create fake pitch drops
          LSTM doesn’t think these are part of a melody'''


    # if use_masking:
    #     x = layers.Masking(mask_value=0.0)(x)

    # Bidirectional LSTM
    x = layers.Bidirectional(layers.LSTM(
                                          lstm_units,
                                          return_sequences=False,   # we only need final embedding
                                          use_cudnn=False
                                          ),name="bilstm"
                            )(x)

    # dropout regularization
    '''Dropout randomly turns off some neurons during training
        The dataset used is relatively small:
          - Only 957 clips total across 8 ragas
          - High correlation between clips from same song

        Without dropout:
          - Model overfits quickly
          - Memorizes song-specific artifacts
          - Doesn’t generalize to new recordings

        Dropout forces the model to learn:
          - general patterns of pitch movement
          - raga grammar
          - prayogas, not specific singer energy or clip noise

        Dropout is especially important for LSTMs
        Because LSTMs tend to overfit on temporal patterns.
    '''

    x = layers.Dropout(dropout_rate, name="lstm_dropout")(x)

    # Dense compression: LSTM embedding
    '''The Dense layer converts the final LSTM output into a compact 128-dimensional embedding:
        This is important for fusion with the CNN branch.
        The two branches (CNN & LSTM) output different sizes:
          CNN embedding → 256 dimensions
          LSTM embedding → could be 128–256 depending on BiLSTM units'''

    x = layers.Dense(
                      128,
                      activation="relu",
                      name="lstm_dense"
                    )(x)

    #Output Embedding Layer
    lstm_embedding = layers.Dropout(0.2)(x)

    return lstm_input, lstm_embedding
```

## D. Fusion Layer (Concatenate[CNN, LSTM]) & derive output Probabilities

### 1. About Fusion Layer

Fusion layer pipeline is structured as:<br>
[CNN Embedding]<br>
[LSTM Embedding]<br>
⬇️ Concatenate<br>
⬇️ Dropout<br>
⬇️ Dense(256, relu)<br>
⬇️ Dense(num_classes, softmax) ← FINAL OUTPUT<br>
⬇️ Output probabilities<br>
This is a standard multi-branch fusion architecture.

### 2. UDF for building the fusion layer

```python
def build_fusion_layer(cnn_embedding, lstm_embedding, num_classes):
    """
    Fuse CNN and LSTM embeddings and create final softmax output.

    cnn_embedding  : Tensor, shape (None, C1)
    lstm_embedding : Tensor, shape (None, C2)
    num_classes    : number of ragas
    """

    # Concatenate embeddings
    fused = Concatenate(name="fusion_concat")([cnn_embedding, lstm_embedding])  # → (None, C1+C2)

    # Optional regularization
    fused = Dropout(0.3, name="fusion_dropout")(fused)

    # Fusion dense layer (learns joint spectral + pitch features)
    fused = Dense(256, activation="relu", name="fusion_dense")(fused)

    # Final softmax classification layer
    output = Dense(num_classes, activation="softmax", name="output")(fused)

    return output
```

### 3. Final Fusion Model Creation

```python
output = build_fusion_layer(cnn_embedding, lstm_embedding, num_classes)
model = Model(inputs=[cnn_input, lstm_input], outputs=output)
```

With this step, The Input Preparation & model architecture building is completed.

- Prepared CNN input<br>
- Prepared LSTM input<br>
- Built CNN branch<br>
- Built LSTM branch<br>
- Fused them<br>
- Added softmax output<br>
- Model Created

### 2. Callbacks


**About CallBack**<br><br>
Callbacks are essentially functions or objects that the training loop calls automatically during execution. They allow you to insert custom behavior without changing the core model training code (model.fit() in Keras).<br><br>
The main purpose of callback is for Monitoring, control, automation, and intervention during the training loop.<br><br>

Common uses for callbacks include:<br>
*Saving the Model*: Automatically save the model's weights periodically or only when a specific metric (like validation accuracy) improves (Keras ModelCheckpoint callback).<br><br>
*Early Stopping:* Stop the training process automatically if the model performance on a validation set stops improving for a certain number of epochs, preventing overfitting (Keras EarlyStopping callback).<br><br>
*Logging and Visualization:* Log metrics and model graphs to tools like TensorBoard for visual analysis (Keras TensorBoard callback).<br><br>
*Learning Rate Scheduling*: Dynamically change the learning rate during training, which can help the model converge faster or achieve better results (Keras ReduceLROnPlateau callback)

```python
# Includes: EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

callbacks = [
              tf.keras.callbacks.EarlyStopping(
                                                monitor="val_loss",
                                                patience=8,
                                                restore_best_weights=True,
                                                verbose=1
                                             ),

              tf.keras.callbacks.ReduceLROnPlateau(
                                                    monitor="val_loss",
                                                    factor=0.5,
                                                    patience=4,
                                                    min_lr=1e-7,
                                                    verbose=1
                                                  ),

              tf.keras.callbacks.ModelCheckpoint(
                                                  filepath="best_raga_cnn_lstm_model.h5",
                                                  monitor="val_loss",
                                                  save_best_only=True,
                                                  verbose=1
                                                )
          ]
```
